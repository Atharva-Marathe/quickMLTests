```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Tutti-Table 1.csv',names=['Id','R','Nome','Squadra','Pg','Mv','Mf','Gf','Gs','Rp','Rc','R+','R-','Ass','Asf','Amm','Esp','Au'])
df.describe()
```

```python
#plot voto e fantavoto for each roma player
roma_index = np.asarray(np.where((df['Squadra']=='Roma') & (df['Pg']>10)))
roma_scores = df['Mv'][roma_index[0]]
roma_players = df['Nome'][roma_index[0]]
roma_scores_fanta = df['Mf'][roma_index[0]]


plt.figure(figsize=(19, 3))
plt.scatter(roma_players,roma_scores)
plt.scatter(roma_players,roma_scores_fanta)
plt.xticks(rotation=45)
plt.show()
```
